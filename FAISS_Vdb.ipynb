```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import DirectoryLoader,PyPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceInstructEmbeddings
import os
```

```python
#creating a vectorstore folder to store FAISS vector db embeddings
vectorstore_path ="/content/drive/MyDrive/tourism in india RAG"
db_faiss_path = os.path.join(vectorstore_path,"db_faiss")

#Create Directories
os.makedirs(db_faiss_path,exist_ok=True)

# Verify Directory
os.path.exists(vectorstore_path), os.path.exists(db_faiss_path)

Data_path = "/content/drive/MyDrive/tourism in india RAG/india reports"
```

Load data from pdfs

```python
%%time
loader = DirectoryLoader(Data_path,
                          glob='**/*.pdf',
                          loader_cls=PyPDFLoader, show_progress=True,
                          use_multithreading=True
                         )

documents = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=300,
                                                chunk_overlap=50)
texts = text_splitter.split_documents(documents)
```

```python
print(len(documents))
documents[0]
```

# Creating Vector Embedding

```python
!pip install -U langchain-huggingface --qq
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
import sentence_transformers
```

```python
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2', model_kwargs={'device': 'cuda'})

db = FAISS.from_documents(texts, embeddings)
db.save_local(db_faiss_path)
print("Saved into the vector database")
```
